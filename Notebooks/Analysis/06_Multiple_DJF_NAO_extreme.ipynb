## Isolating and combining NAO extremes for multiple DJF seasons

Here we isolate members that had extreme NAO conditions in 2nd, 3rd or 4th DJF season. We then combine them while maintining the position of NAO extreme position. This gives better accuracy in terms of statistical significance and averaging. 

In another test, we isolate members that had extreme NAO onciditions in two consecutive DJF seasons.

```python
ppdir = "/home/users/hkhatri/DePreSys4_Data/Data_Anomaly_Compute/"

ds_TS = []

var_list2 = ['hfds', 'tos', 'sos']

for var in var_list2:
    
    ds1 = []
    
    for i in range(0,10):
    
        d = xr.open_dataset(ppdir + "Area_mean_timeseries/Anomaly_" + var + "_r" + str(i+1) + ".nc")
        
        ds1.append(d)
    
    ds1 = xr.concat(ds1, dim='r')
    
    ds_TS.append(ds1)

ds_TS = xr.concat(ds_TS, dim='var')

ds_mask = xr.open_dataset('/home/users/hkhatri/DePreSys4_Data/Data_Consolidated/Mask_Regions.nc')
ds_area = xr.open_dataset('/home/users/hkhatri/DePreSys4_Data/Data_Consolidated/Area.nc')
ds_mask['area'] = ds_area['area_t'].rename({'x':'i', 'y':'j'})
```

```python
# Compute mean surface properties north and south of 40 N

region_list = ['Labrador_Sea', 'Irminger_Sea', 'Iceland_Basin', 'North_East_Region', 'South_West_Region', 
               'South_East_Region', 'North_Atlantic']

mask_val = 0.
North_var = []
North_area = []

for region in region_list:
    
    dA = ds_mask['area'].where(ds_mask['mask_regions'] == mask_val).sum(['i', 'j'])
    
    tmp = ds_TS[region] * dA
    
    North_var.append(tmp)
    North_area.append(dA)
    
    mask_val = mask_val + 1
    
    
ds_TS['North_40N'] = (sum(North_var)/ sum(North_area)).compute()

dA_NorAtl = ds_mask['area'].where(ds_mask['mask_North_Atl'] == 0.).sum(['i', 'j'])
tmp = ds_TS['North_Atlantic'] * dA_NorAtl

ds_TS['South_40N'] = ((tmp - sum(North_var)) / (dA_NorAtl - sum(North_area))).compute()
```

### NAO cases 

Select NAO extreme members that are chosen for representation. These correspond to 2nd, 3rd and 4th DJFs.

```python
NAO_season = ds_NAO['NAO'].copy()
tim = ds_NAO['time_val'].isel(start_year=0).drop('start_year')
NAO_season = NAO_season.assign_coords(time=tim)

NAO_season = NAO_season.isel(time=slice(1,len(NAO_season.time)-1))
NAO_season = NAO_season.resample(time='QS-DEC').mean('time').compute()
```

```python
NAO_cut = 2.5

count_NAOp = 0
count_NAOn = 0

for ind in range(4,13,4):
    
    c_NAOp = (xr.where(NAO_season.isel(time=ind) >= NAO_cut, 1, 0)).sum().values
    c_NAOn = (xr.where(NAO_season.isel(time=ind) <= -NAO_cut, 1, 0)).sum().values
    
    count_NAOp =  count_NAOp + c_NAOp
    count_NAOn =  count_NAOn + c_NAOn

print('NAO positive members = ', count_NAOp)
print('NAO negative members = ', count_NAOn)
```

```python
case_list = ['NAOp', 'NAOn']
NAO_cut = 2.5

grid1 = plt.GridSpec(1, 2, wspace=0.3, hspace=0.3)
fig = plt.figure(figsize=[16,4])

ax = fig.add_subplot(grid1[0, 0])
for year in range(0,len(ds_NAO['start_year'])):
    
    for r in range(0,10):
        
        ax.plot(np.arange(0,len(NAO_season.time))/4, NAO_season.isel(start_year=year, r=r),'o', alpha=0.5)
        
ax.grid(color='gray', alpha=0.5, linestyle='--', lw=2.)
ax.set_ylim([-5, 5])
ax.set_ylabel('NAO', fontsize=14)
ax.set_xlabel('Year', fontsize=14)


ax = fig.add_subplot(grid1[0, 1])
for case in case_list:
    
    tmp = []
    if (case == 'NAOp'):
        
        N = count_NAOp
        std_err = (1.96/np.sqrt(np.ceil(N)))
            
        for ind in range(4,13,4):

            tmp1 = ds_NAO['NAO'].where(NAO_season.isel(time=ind).drop('time') >= NAO_cut)
            tmp1 = tmp1.isel(time = slice((int(ind/4)-1)*12, (int(ind/4) + 7)*12 + 5))

            tmp.append(tmp1)

        tmp = xr.concat(tmp,dim='case')
        tmp = tmp.rolling(time=6, center=True).mean('time')
        tmp_std = tmp.std(['start_year','r', 'case'])
        tmp = tmp.sum(['start_year','r','case']) / N
        

    elif(case == 'NAOn'):
        
        N = count_NAOn
        std_err = (1.96/np.sqrt(np.ceil(N)))
            
        for ind in range(4,13,4):

            tmp1 = ds_NAO['NAO'].where(NAO_season.isel(time=ind).drop('time') <= -NAO_cut)
            tmp1 = tmp1.isel(time = slice((int(ind/4)-1)*12, (int(ind/4) + 7)*12 + 5))

            tmp.append(tmp1)

        tmp = xr.concat(tmp,dim='case')
        tmp = tmp.rolling(time=6, center=True).mean('time')
        tmp_std = tmp.std(['start_year','r', 'case'])
        tmp = tmp.sum(['start_year','r','case']) / N
                      
    ax.plot(np.arange(0+10,len(tmp.time)+10) / 12 - 2, tmp, lw=3.)
    ax.fill_between(np.arange(0+10,len(tmp.time)+10) / 12 - 2, (tmp.values - std_err*tmp_std.values), 
                    (tmp.values + std_err*tmp_std.values), alpha=0.3)
    ax.axhline(y=0, ls='-.', lw=0.5, color='k')
    ax.grid(color='gray', alpha=0.5, linestyle='--', lw=2.)
    ax.set_ylabel('NAO', fontsize=14)
    ax.set_xlabel('Year', fontsize=14)
```

# Members with consecutive NAO extremes

Consecuive NAO extremes (2nd and 3rd DJF or 3rd and 4th DJF and so on)
