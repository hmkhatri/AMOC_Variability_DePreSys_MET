## Check combining files

This to check that individual files on ceda server are combined into a single dataset properly, the runs on the full dataset return no.

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.crs as ccrs
import cartopy.feature as cfeature

import warnings
warnings.filterwarnings('ignore')
```

### Check for mean drift data

```python
ppdir1="/home/users/hkhatri/DePreSys4_Data/Data_Drift_Removal/Drift_2016_DCPP/"

var = 'tos'

ds_drift = []
    
for lead_year in range(0, 11):
    
    ds1 =[]
    
    for r in range(1,11):
    
        d = xr.open_dataset(ppdir1 + "Drift_" + var + "_r" + str(r) + "_Lead_Year_" + str(lead_year + 1) + ".nc", decode_times= False)
        ds1.append(d)
        
    ds1 = xr.concat(ds1, dim='r')
    
    ds_drift.append(ds1)
        
ds_drift = xr.concat(ds_drift, dim='lead_year')
```

```python
print(ds_drift)
```

### Check for full data

```python
ppdir="/badc/cmip6/data/CMIP6/DCPP/MOHC/HadGEM3-GC31-MM/dcppA-hindcast/"

ds = []

for year in range(2011, 2013, 1):

    ds1 = []

    for i in range(0,4):
        
        var_path = "s" + str(year) +"-r" + str(i+1) + "i1p1f2/Omon/" + var + "/gn/files/d20200417/"
        d = xr.open_mfdataset(ppdir + var_path + "*.nc")
        ds1.append(d)
        
    ds1 = xr.concat(ds1, dim='r')
    ds1['time_val'] = d['time'] # dropping time coordinate is required; otherwise xarray concat does not work properly
    ds1 = ds1.drop(['time', 'vertices_latitude', 'vertices_longitude', 'time_bnds'])
    
    ds.append(ds1)
    
ds = xr.concat(ds, dim='year')
```

```python
print(ds)
```

### Compute DJF seasonal anomaly 

```python
ds = ds.assign_coords(time=ds['time_val'].isel(year=0)) # this is just to get seasonal averaging work
ds = ds.isel(i=slice(749,1199), j = slice(699, 1149))

tos_resam = ds.resample(time='QS-DEC').mean('time')
tos_resam = tos_resam.sel(time = tos_resam['time.season'] == 'DJF')

# again drop time and change names to be consistent with ds_drift
tos_resam = tos_resam.drop('time')
tos = tos_resam.rename({'time':'lead_year'})

tos = tos.transpose('year','lead_year','r','j','i')
```

```python
print(tos)
```

```python
tos_anom = tos - ds_drift.isel(r=slice(0,4))

print(tos_anom)
```

```python
tos_anom['tos'].isel(year=0, r=0, lead_year=0).plot()
```

```python
ds_mask = xr.open_dataset('/home/users/hkhatri/DePreSys4_Data/Data_Consolidated/Mask_Regions.nc')

ds_area = xr.open_dataset('/home/users/hkhatri/DePreSys4_Data/Data_Consolidated/Area.nc')

ds_mask['area'] = ds_area['area_t'].rename({'x':'i', 'y':'j'})

print(ds_mask)
```

```python
def Compute_area_weighted_mean(ds, area, mask, lat = [0., 90.]): 
    
    #tmp = ds.where(mask).sel(lat=slice(lat[0], lat[1]))
    #dA = area.where(mask).sel(lat=slice(lat[0], lat[1]))
    
    tmp1 = (ds * area).sum(['i', 'j']) / area.sum(['i', 'j']) 
    
    return tmp1

SST_Anom = Compute_area_weighted_mean(tos_anom.where(ds_mask['mask_North_Atl'] == 0.), 
                                       ds_mask['area'], ds_mask, lat = [0., 80.])

SST_drift = Compute_area_weighted_mean(ds_drift['tos'].where(ds_mask['mask_North_Atl'] == 0.), 
                                       ds_mask['area'], ds_mask, lat = [0., 80.])
```

```python
SST_Anom = SST_Anom.compute()

SST_drift = SST_drift.compute()

print(SST_Anom)
```

```python
grid = plt.GridSpec(1, 3, wspace=0.2, hspace=0.4)
fig = plt.figure(figsize=[20,4])

ax = fig.add_subplot(grid[0, 0])

for i in range(0,4):
    ax.plot(SST_Anom['lead_year'] + 2011, SST_Anom.isel(r=i, year=0))
    ax.set_ylabel(r'SST Anomaly (C)', fontsize=14)
    ax.set_xlabel(' ')
    ax.set_title('North Atlantic', fontsize=14)
    ax.grid()
    plt.yticks(fontsize = 12)
    plt.xticks(fontsize = 12)

ax = fig.add_subplot(grid[0, 1])
for i in range(0,4):
    ax.plot(SST_Anom['lead_year'] + 2012, SST_Anom.isel(r=i, year=1))
    ax.set_ylabel(r'SST Anomaly (C)', fontsize=14)
    ax.set_xlabel(' ')
    ax.set_title('North Atlantic', fontsize=14)
    ax.grid()
    plt.yticks(fontsize = 12)
    plt.xticks(fontsize = 12)
    
ax = fig.add_subplot(grid[0, 2])
for year in range(0,2):
    ax.plot(SST_Anom['lead_year'] + year, SST_Anom.isel(year=year).mean('r'), lw=3)
    ax.set_ylabel(r'SST Anomaly (C)', fontsize=14)
    ax.set_xlabel(' ')
    ax.set_title('North Atlantic', fontsize=14)
    ax.grid('on')
    plt.yticks(fontsize = 12)
    plt.xticks(fontsize = 12)
```
