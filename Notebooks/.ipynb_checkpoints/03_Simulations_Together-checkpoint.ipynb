## Combining data for all simulations (1960 - 2016)

In this notebook, we combine data for all simulations to make sure that all datafiles can be accessed within a single xarray dataset.

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

### Check data for overturning strength and transport diagnostics  

```python
ppdir="/home/users/hkhatri/DePreSys4_Data/Data_Consolidated/"

# variable list to keep in the dataset
#var_list = ['hfbasin_atlantic', 'hfbasinpmdiff_atlantic', 'hfovgyre_atlantic', 'hfovovrt_atlantic', 'sophtadv_atlantic', 
#            'sltbasin_atlantic', 'sltbasinpmdiff_atlantic', 'sltovgyre_atlantic', 'sltovovrt_atlantic', 'sopstadv_atlantic',
#            'zomsfatl', 'time']

var_list = ['hfbasin_atlantic', 'sltbasin_atlantic', 'zomsfatl', 'zosalatl','zosrfatl','zotematl','time',
           'deptht_bounds', 'depthw_bounds']

ds = []

stride_year = 1

# Loop over year to combine indivual year files
for year in range(1960, 2017 , stride_year):
    
    d = xr.open_dataset(ppdir + str(year) + "_diaptr.nc", decode_times= False, chunks={'r':1})
    d = d.get(var_list)
    d = d.drop('time')
    ds.append(d)
    
ds = xr.concat(ds, dim='start_year')

ds = ds.assign(start_year = np.arange(1960, 2017, stride_year))

ds = ds.chunk({'start_year': 1})

# this is get back time units properly
# reading files without decode_times = False gives an error

ds = xr.decode_cf(ds)

print(ds)
```

### Overturning Strength

We first compute yearly means for all simulations and across all ensembles. The time series of overturning strength is plotted for each simulation corresponding to years between 1960 and 2016. Shading represents the 95% confidence interval computed for 10 ensemble runs for each simulation year.

```python
# Check how Overturning strength looks like at 30 N, 40 N and 50 N

psi_max = ds.zomsfatl.mean(['x'])
psi_max = psi_max.assign_coords(y=ds.nav_lat.isel(x=0))
psi_max = psi_max.sel(y=[30., 50.], method="nearest")
lat = ['30 N', '50 N']
```

```python
N = 10 # Ensembles
ci = 2.262 # coefficiet for N = 10 (mu = N -1) and 95% confidence interval -> Data Analysis Methods book
std_err = (ci/np.sqrt(np.ceil(N)))

grid = plt.GridSpec(2, 1, wspace=0.4, hspace=0.4)
fig = plt.figure(figsize=[22,8])

col = plt.cm.Dark2(np.linspace(0.,0.9, len(ds.start_year)))

for i in range(0,2):
    
    ax = fig.add_subplot(grid[i, 0])
    
    for j in range(0,len(ds.start_year)):
        
        tmp = (psi_max.isel(y=i, start_year=j, time_counter=slice(2,122))).max(dim='depthw').groupby('time_centered.year').mean('time_counter')
        ax.plot(tmp.year, tmp.mean('r'), lw = 2., color=col[j])
        ax.fill_between(tmp.year, tmp.mean('r') - std_err*tmp.std('r'), tmp.mean('r') + std_err*tmp.std('r'), 
                        facecolor=col[j], alpha=0.3)
        
    ax.set_ylabel(r'Overturning Strength (Sv)', fontsize=14)
    #ax.set_xlabel('Time', fontsize=14)
    plt.title(lat[i], fontsize=14)
    ax.grid()
    plt.yticks(fontsize = 12)
    plt.xticks(fontsize = 12)
```

### Heat and Salt Anomalies

We compute heat and salt anomly treds (in the upper 1300 m) and comapre against Williams et al. (2015).

$$Q' = \rho_oC_p \int_{-z}^{0}\theta'(x,y,z,t)dz$$
$$S' = \int_{-z}^{0}s'(x,y,z,t)dz$$

```python
dz = ds['deptht_bounds'].isel(start_year=0, r=0, time_counter=0).diff('axis_nbounds')

rho = 1035.0
cp = 4185.5

# compute time-mean T and S

T_mean = ds['zotematl'].mean(['start_year','time_counter', 'r', 'x'])
S_mean = ds['zosalatl'].mean(['start_year','time_counter', 'r', 'x'])

T_mean = T_mean.assign_coords(y=ds.nav_lat.isel(x=0))
S_mean = S_mean.assign_coords(y=ds.nav_lat.isel(x=0))

Q = rho * cp * T_mean
```

```python
Q.where(Q.deptht <= 1300.).plot()
```

```python
T = ds['zotematl'].assign_coords(y=ds.nav_lat.isel(x=0))
S = ds['zosalatl'].assign_coords(y=ds.nav_lat.isel(x=0))

col = plt.cm.Dark2(np.linspace(0.,0.9, len(ds.start_year)))

grid = plt.GridSpec(2, 2, wspace=0.2, hspace=0.2)
fig = plt.figure(figsize=[22,8])

for i in range(0,2):
    ax = fig.add_subplot(grid[i, 0])

    for j in range(0,len(ds.start_year)):
        
        if(i==0):
            tmp = (T.isel(start_year=j, time_counter=slice(2,122), 
                          x=0).sel(y=slice(5.,46.)).mean(['r','y']).groupby('time_centered.year').mean('time_counter'))
            tmp = ((tmp * rho * cp - Q.sel(y=slice(5.,46.)).mean('y')) * dz).where(tmp.deptht <= 1300.).sum('deptht')
                   
        if(i==1):
            tmp = (T.isel(start_year=j, time_counter=slice(2,122), 
                          x=0).sel(y=slice(46.,75.)).mean(['r','y']).groupby('time_centered.year').mean('time_counter'))
            tmp = ((tmp * rho * cp - Q.sel(y=slice(46., 75.)).mean('y')) * dz).where(tmp.deptht <= 1300.).sum('deptht')
                   
        ax.plot(tmp.year, tmp, lw = 2., color=col[j])
        ax.set_ylabel(r'$Q^{\prime}$ upper 1300 m (J/m$^2$)', fontsize=14)
        ax.axhline(y=0.0, color="black", linestyle="--") 
        if(i==0):
            plt.title('5 N - 46 N', fontsize=14)
        else:
            plt.title('46 N - 75 N', fontsize=14)
        
    ax = fig.add_subplot(grid[i, 1])

    for j in range(0,len(ds.start_year)):
        
        if(i==0):
            tmp = (S.isel(start_year=j, time_counter=slice(2,122), 
                          x=0).sel(y=slice(5.,46.)).mean(['r','y']).groupby('time_centered.year').mean('time_counter'))
            tmp = (((tmp - S_mean.sel(y=slice(5.,46.)).mean('y')) * dz).where(tmp.deptht <= 1300.).sum('deptht') /
                      dz.where(tmp.deptht <= 1300.).sum('deptht'))
                   
        if(i==1):
            tmp = (S.isel(start_year=j, time_counter=slice(2,122), 
                          x=0).sel(y=slice(46.,75.)).mean(['r','y']).groupby('time_centered.year').mean('time_counter'))
            tmp = (((tmp - S_mean.sel(y=slice(46., 75.)).mean('y'))* dz).where(tmp.deptht <= 1300.).sum('deptht') /
                      dz.where(tmp.deptht <= 1300.).sum('deptht'))
                   
        ax.plot(tmp.year, tmp, lw = 2., color=col[j])
        ax.set_ylabel('$S^{\prime}$ upper 1300 m (psu)', fontsize=14)
        ax.axhline(y=0.0, color="black", linestyle="--") 
        if(i==0):
            plt.title('5 N - 46 N', fontsize=14)
        else:
            plt.title('46 N - 75 N', fontsize=14)    
            
```

#### Compare heat content anomaly and salt anomaly

```python
dz = ds['deptht_bounds'].isel(start_year=0, r=0, time_counter=0).diff('axis_nbounds')
dz = dz.isel(axis_nbounds = 0)

rho = 1035.0
cp = 4185.5

# compute anomaly T and S

ds['time'] = ds['time_centered'].isel(start_year=0)
ds = ds.assign_coords(time_counter=ds.time)

T_resam = ds['zotematl'].resample(time_counter='QS-DEC').mean('time_counter')
T_resam = T_resam.sel(time_counter = T_resam['time_counter.season'] == 'DJF').mean(['x'])
T_resam = T_resam.assign_coords(y=ds.nav_lat.isel(x=0))
T_resam = T_resam.drop('time_counter').rename({'time_counter': 'lead_year'})

T1 = ds_drift['zotematl'].mean(['x'])
T1 = T1.assign_coords(y=ds_drift.nav_lat.isel(x=0))

T_anom = T_resam.transpose('start_year', 'lead_year','r','y','deptht') - T1

Q_anom = ((rho * cp * T_anom * dz).where(T_anom.deptht <= 500.).sum('deptht')).compute()

S_resam = ds['zosalatl'].resample(time_counter='QS-DEC').mean('time_counter')
S_resam = S_resam.sel(time_counter = S_resam['time_counter.season'] == 'DJF').mean(['x'])
S_resam = S_resam.assign_coords(y=ds.nav_lat.isel(x=0))
S_resam = S_resam.drop('time_counter').rename({'time_counter': 'lead_year'})

S1 = ds_drift['zosalatl'].mean(['x'])
S1 = S1.assign_coords(y=ds_drift.nav_lat.isel(x=0))

S_anom = S_resam.transpose('start_year', 'lead_year','r','y','deptht') - S1
S_anom = (S_anom* dz).where(S_anom.deptht <= 500.).sum('deptht') / dz.where(S_anom.deptht <= 500.).sum('deptht')
S_anom = S_anom.compute()
```
